Importing dependencies:

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading dataset to pandas dataframe
credit_data=pd.read_csv('/content/creditcard.csv')
```

```python
# 1st 5 rowrs:
credit_data.head()
```

time: Number of seconds elapsed between this transaction and the first transaction in the dataset

It contains only numerical input variables which are the result of a PCA transformation. 


[PCA:Principal Component Analysis is an unsupervised learning algorithm that is used for the dimensionality reduction in machine learning. It is a statistical process that converts the observations of correlated features into a set of linearly uncorrelated features with the help of orthogonal transformation.]

The feature 'Amount' is the transaction Amount.

Feature 'Class' is the response variable and it takes value 1 in case of fraud and 0 otherwise.

Under sampling

Build a sample dataset containing similar distribution of normal and frudulent transaction: 

```python
# number of fraud transactions = 271
```

```python
legit_sample=legit.sample(n=271)
```

Concatenating two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.shape
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
# now we get unfiformly distributed data
```

```python
new_dataset.groupby('Class').mean()
```

```python
# we see the same mean as unbalanced data distribution 
```

Splitting the data into features and target

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
x
```

```python
y
```

Splitting the data into train and test

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
x.shape
```

```python
print('xtrain.shape:',xtrain.shape)
print('xtest.shape:',xtest.shape)
print('ytrain.shape:',ytrain.shape)
print('ytest.shape:',ytest.shape)
```

Model training

Logistic Regression:

```python
model=LogisticRegression()
```

```python
# training the logistic regression model with training data
```

```python
model.fit(xtrain,ytrain)
```

Model Evaluation

Acurracy score:


```python
# accuracy on training data:
xtrain_prediction=model.predict(xtrain)
```

```python
xtrain_prediction
```

```python
train_data_acc=accuracy_score(xtrain_prediction,ytrain)
print('train_data_acc:',train_data_acc)
```

```python
# test data accuracy:
xtest_prediction=model.predict(xtest)
```

```python
test_data_acc=accuracy_score(xtest_prediction,ytest)
print('train_data_acc:',test_data_acc)
```
